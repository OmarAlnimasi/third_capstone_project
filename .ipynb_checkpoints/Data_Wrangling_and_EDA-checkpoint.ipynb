## Exploratory Data Analysis:

We will start the Exploratory Data Analysis (EDA) step by looking at the correlation matrix, but first we need to change our target column into numerical

```python
df['Attrition_Flag'].replace({'Existing Customer':1,'Attrited Customer':0},inplace=True)
df['Attrition_Flag'].astype('int64')
df.rename(columns={'Attrition_Flag':'Existing_Customer'},inplace=True)
```

```python
corr = df.corr()
```

```python
fig, ax = plt.subplots(figsize=(15,12))
sns.heatmap(corr)
```

As we can see here, the most correlated columns are Total_Trans_Ct, Total_Ct_Chng_Q4_Q1, and Abg_Utilization_Ratio, while we can see some other columns that are anticorreleted, like Contacts_Count_12_mon, which is an unusal relation.
So we can break it down in terms of pair plots, and see the numerical values represented vs Existing_Members

```python
df_n=df[numerical]
sns.pairplot(df_n)
```

These figures are telling the same story we are getting from the heatmap, Since we only have two catagories in the target column, we cannot get much from boxplots as we will get minimal information from the figures. So we need to extract the data now for pre-processing and modeling 

```python
df.to_csv('data2.csv', header=True, index=False)
```
